```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
# Download required NLTK data
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
```

```python
def perform_lemmatization():
    text = "running ran runs easily possibly happiness happy jumping jumped better worse"
    
    words = word_tokenize(text)
    print("Original words:", words)
    
    # Initialize lemmatizer
    lemmatizer = WordNetLemmatizer()
    
    # Lemmatize without POS context
    simple_lemmas = [lemmatizer.lemmatize(word) for word in words]
    
    # Lemmatize with POS context
    pos_lemmas = []
    for word in words:
        # Try different POS tags for better lemmatization
        lemma = lemmatizer.lemmatize(word, pos='v')  # Try verb first
        if lemma == word:  # If no change, try noun
            lemma = lemmatizer.lemmatize(word, pos='n')
        if lemma == word:  # If still no change, try adjective
            lemma = lemmatizer.lemmatize(word, pos='a')
        if lemma == word:  # If still no change, try adverb
            lemma = lemmatizer.lemmatize(word, pos='r')
        pos_lemmas.append(lemma)
    
    print("\nLemmatization Results:")
    print(f"{'Word':<15} {'Simple Lemma':<15} {'POS Lemma':<15}")
    print("-" * 45)
    
    for word, simple, pos in zip(words, simple_lemmas, pos_lemmas):
        print(f"{word:<15} {simple:<15} {pos:<15}")
```

```python
if __name__ == "__main__":
    perform_lemmatization()
```
