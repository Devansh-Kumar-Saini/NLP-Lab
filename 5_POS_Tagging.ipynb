```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
# Download required NLTK data
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
```

```python
def perform_pos_tagging():
    text = "The quick brown fox jumps over the lazy dog. She is reading a book beautifully."
    
    # Tokenize the text
    words = word_tokenize(text)
    print("Tokens:", words)
    
    # Perform POS tagging
    pos_tags = pos_tag(words)
    
    print("\nPOS Tagging Results:")
    print(f"{'Word':<15} {'POS Tag':<10} {'Description':<20}")
    print("-" * 50)
    
    # POS tag descriptions
    tag_dict = {
        'NN': 'Noun, singular',
        'NNS': 'Noun, plural',
        'VB': 'Verb, base form',
        'VBD': 'Verb, past tense',
        'VBG': 'Verb, gerund',
        'VBN': 'Verb, past participle',
        'VBP': 'Verb, non-3rd person',
        'VBZ': 'Verb, 3rd person',
        'JJ': 'Adjective',
        'JJR': 'Adjective, comparative',
        'JJS': 'Adjective, superlative',
        'RB': 'Adverb',
        'RBR': 'Adverb, comparative',
        'RBS': 'Adverb, superlative',
        'DT': 'Determiner',
        'IN': 'Preposition',
        'PRP': 'Personal pronoun',
        'PRP$': 'Possessive pronoun'
    }
    
    for word, tag in pos_tags:
        description = tag_dict.get(tag, 'Other')
        print(f"{word:<15} {tag:<10} {description:<20}")
```

```python
if __name__ == "__main__":
    perform_pos_tagging()
```
