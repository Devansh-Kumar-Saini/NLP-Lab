```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
# Download required NLTK data
nltk.download('punkt')
```

```python
def tokenize_text():
    text = """
    Natural Language Processing (NLP) is a fascinating field. 
    It helps computers understand human language. 
    Tokenization is the first step in NLP pipelines.
    """
    
    # Sentence tokenization
    sentences = sent_tokenize(text)
    print("Sentence Tokenization:")
    for i, sentence in enumerate(sentences, 1):
        print(f"Sentence {i}: {sentence.strip()}")
    
    # Word tokenization
    words = word_tokenize(text)
    print("\nWord Tokenization:")
    print(words)
    
    # Word tokenization for each sentence
    print("\nWord Tokenization per Sentence:")
    for i, sentence in enumerate(sentences, 1):
        words_in_sentence = word_tokenize(sentence)
        print(f"Sentence {i}: {words_in_sentence}")
```

```python
if __name__ == "__main__":
    tokenize_text()
```
