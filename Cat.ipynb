```python
import nltk
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize
```

```python
# Download required NLTK data
nltk.download('punkt')
```

```python
# Example corpus - collection of similar sentences
corpus = [
    "This is my cat",
    "This is my dog",
    "This is a cat",
    "That is my cat",
    "This cat is cute",
    "My cat is sleeping",
    "This is the cat"
]
```

```python
def calculate_sentence_probability(sentence, corpus):
    # Tokenize the corpus
    tokenized_corpus = [word_tokenize(sent.lower()) for sent in corpus]
    
    # Create a flat list of all words
    all_words = [word for sent in tokenized_corpus for word in sent]
    
    # Calculate word frequencies
    freq_dist = FreqDist(all_words)
    
    # Calculate total words in corpus
    total_words = len(all_words)
    
    # Tokenize the input sentence
    input_words = word_tokenize(sentence.lower())
    
    # Calculate probability (using simple unigram model)
    probability = 1.0
    for word in input_words:
        word_prob = freq_dist[word] / total_words
        probability *= word_prob
    
    return probability
```

```python
# Test the probability
sentence = "This is my cat"
prob = calculate_sentence_probability(sentence, corpus)

print(f"Probability of '{sentence}': {prob}")

# Display word frequencies
print("\nWord frequencies in corpus:")
tokenized_sentence = word_tokenize(sentence.lower())
freq_dist = FreqDist([word for sent in corpus for word in word_tokenize(sent.lower())])
for word in tokenized_sentence:
    print(f"'{word}': {freq_dist[word]} occurrences")
```
