```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.chunk import RegexpParser
```

```python
# Download required NLTK data
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
```

```python
def perform_chunking():
    text = "The quick brown fox jumps over the lazy dog near the big river."
    
    # Tokenize and POS tag
    words = word_tokenize(text)
    pos_tags = pos_tag(words)
    
    print("POS Tags:", pos_tags)
    
    # Define chunk grammar
    grammar = r"""
        NP: {<DT>?<JJ>*<NN.*>+}    # Noun Phrase: determiner + adjectives + nouns
        VP: {<VB.*><NP|PP>?}        # Verb Phrase: verb + optional noun/preposition phrase
        PP: {<IN><NP>}              # Preposition Phrase: preposition + noun phrase
    """
    
    # Create chunk parser
    chunk_parser = RegexpParser(grammar)
    
    # Parse the sentence
    result = chunk_parser.parse(pos_tags)
    
    print("\nChunking Result:")
    print(result)
    
    # Visualize the parse tree
    print("\nVisual Representation:")
    result.draw()
```

```python
if __name__ == "__main__":
    perform_chunking()
```
