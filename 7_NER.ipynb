```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk
```

```python
# Force download to a known location
nltk.download('maxent_ne_chunker', download_dir='C:/nltk_data')
nltk.download('words', download_dir='C:/nltk_data')
nltk.download('punkt', download_dir='C:/nltk_data')
nltk.download('averaged_perceptron_tagger', download_dir='C:/nltk_data')

# Add this path to NLTK's search paths
nltk.data.path.append('C:/nltk_data')
```

```python
def perform_ner():
    text = """
    Apple Inc. is headquartered in Cupertino, California. 
    Tim Cook is the CEO of Apple. 
    Microsoft was founded by Bill Gates in Redmond, Washington.
    Elon Musk leads Tesla and SpaceX from Texas.
    """
    
    # Tokenize and POS tag
    words = word_tokenize(text)
    pos_tags = pos_tag(words)
    
    print("POS Tags:", pos_tags)
    
    # Perform Named Entity Recognition
    ner_tree = ne_chunk(pos_tags)
    
    print("\nNER Result:")
    print(ner_tree)
    
    # Extract named entities
    print("\nNamed Entities:")
    entities = []
    for chunk in ner_tree:
        if hasattr(chunk, 'label'):
            entity_text = ' '.join(c[0] for c in chunk)
            entity_type = chunk.label()
            entities.append((entity_text, entity_type))
            print(f"Entity: {entity_text:<25} Type: {entity_type}")
    
    return entities
```

```python
if __name__ == "__main__":
    perform_ner()
```
