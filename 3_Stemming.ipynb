```python
import nltk
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
from nltk.tokenize import word_tokenize
```

```python
def perform_stemming():
    text = "running runner runs easily possibly happiness happy jumping jumper"
    
    words = word_tokenize(text)
    print("Original words:", words)
    
    # Porter Stemmer
    porter = PorterStemmer()
    porter_stems = [porter.stem(word) for word in words]
    
    # Lancaster Stemmer
    lancaster = LancasterStemmer()
    lancaster_stems = [lancaster.stem(word) for word in words]
    
    # Snowball Stemmer (supports multiple languages)
    snowball = SnowballStemmer('english')
    snowball_stems = [snowball.stem(word) for word in words]
    
    print("\nStemming Results:")
    print(f"{'Word':<15} {'Porter':<15} {'Lancaster':<15} {'Snowball':<15}")
    print("-" * 60)
    
    for word, p_stem, l_stem, s_stem in zip(words, porter_stems, lancaster_stems, snowball_stems):
        print(f"{word:<15} {p_stem:<15} {l_stem:<15} {s_stem:<15}")
```

```python
if __name__ == "__main__":
    perform_stemming()
```
