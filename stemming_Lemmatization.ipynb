```python


# Sample text
text = "A major limitation of Part-of-Speech (POS) tagging in NLP is ambiguity, where a single word can have multiple POS tags depending on the context. This ambiguity can lead to incorrect tagging if not resolved by context. For example, the word 'book' can be a noun (as in, 'a book') or a verb (as in, 'to book a flight')"

# Tokenize the text
tokens = word_tokenize(text)

# Initialize the stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Perform stemming
stemmed_words = [stemmer.stem(word) for word in tokens]

# Perform lemmatization
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in tokens]  # 'v' is for verb, change based on need

print("Original Text:", text)
print("\nStemmed Words:", stemmed_words)
print("\nLemmatized Words:", lemmatized_words)
```
